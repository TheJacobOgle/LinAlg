```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Python List Version

# 2D Vector
v2 = [3, -2]
print("V2: ", v2)

# 3D Vector
v3 = [4, -3, 1]
print("V3: ", v3)

# Row to column vector
v3t = np.transpose(v3) # Will not work since the v3 is not np array
print("V3T: ", v3t)
```

```python
plt.style.use('dark_background')  # Switch to dark mode

plt.plot([0, v2[0]], [0, v2[1]])
plt.axis('equal')
plt.plot([-4,4], [0,0], 'k--')
plt.plot([0,0], [-4,4], 'k--')
plt.grid()
plt.axis((-4,4,-4,4))
plt.show()

fig = plt.figure(figsize=plt.figaspect(1))
ax = fig.add_subplot(111, projection='3d')
ax.plot([0, v3[0]], [0, v3[1]], [0, v3[2]])

plt.show()
```

```python
# Addition and Subtraction

v1 = np.array([1,2,3])
v2 = np.array([10,11,12])

v3p = v1 + v2
print(f"v3plus: {v3p}")

v3m = v1 - v2
print(f"v3minus: {v3m}")

plt.plot([0, v1[0]],[0, v1[1]],'b',label='v1')
plt.plot([0, v2[0]]+v1[0],[0, v2[1]]+v1[1],'r',label='v2')
plt.plot([0, v3[0]],[0, v3[1]],'k',label='v1+v2', c="g")

plt.legend()
plt.axis('square')
plt.axis((-10, 10, -10, 10 ))
plt.grid()
plt.show()
```

```python
# Vector Scalar Multiplication
v1 = np.array([3,2])
l = -0.3 # lambda scalar vector
w = l * v1
print(w)

plt.plot([0, v1[0]],[0, v1[1]],'b',label='$v_1$')
plt.plot([0, w[0]],[0, w[1]],'r:',label='$\lambda v_1$')

plt.legend()
plt.axis('square')
axlim = max([max(abs(v1)),max(abs(w))])*1.5 # dynamic axis lim
plt.axis((-axlim,axlim,-axlim,axlim))
plt.grid()
plt.show()
```

```python
# Dot Product

v = np.array([1,2,3,4,5,6,7])
w = np.array([12,3,1,5,10,11,7])
print(v.shape, w.shape)

# Methods
dp1 = sum(np.multiply(v, w))

dp2 = np.dot(v,w)

dp3 = np.matmul(v, w)

print(dp1, dp2, dp3)
```

```python
# Dot Product is Distributive
n = 10
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)

# the two results
res1 = np.dot( a , (b+c) )
res2 = np.dot(a,b) + np.dot(a,c)

print([ res1,res2 ])

# Dot Product is Commutative
res1 = np.dot(v, w)
res2 = np.dot(w, v)
print([ res1,res2 ])
```
